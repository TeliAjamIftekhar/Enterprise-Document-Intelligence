```python
import json

with open("../data/pages.json", "r", encoding="utf-8") as f:

    pages = json.load(f)

print("Pages Loaded:", len(pages))
```

Let's inspect the document.

```python
for page in pages[:15]:

    print(
        f"Page {page['page']:3d} | Characters = {len(page['text'])}"
    )
```

Cell 3 — Remove Empty Pages

```python
clean_pages = []

for page in pages:

    if len(page["text"].strip()) > 100:

        clean_pages.append(page)

print("Original Pages :", len(pages))
print("Useful Pages   :", len(clean_pages))
```

Cell 4 — Save Clean Pages

```python
from pathlib import Path

output_path = Path("../data/clean_pages.json")

with open(output_path, "w", encoding="utf-8") as f:

    json.dump(
        clean_pages,
        f,
        indent=2,
        ensure_ascii=False
    )

print("Saved:", output_path)
```
