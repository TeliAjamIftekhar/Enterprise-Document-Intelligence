Cell 1

Load cleaned pages.

```python
import json
import os
```

Cell 2 — Load clean_pages.json

```python
INPUT_FILE = "../data/clean_pages.json"

with open(INPUT_FILE, "r", encoding="utf-8") as f:
    pages = json.load(f)

print("="*60)
print("Pages Loaded :", len(pages))
```

Cell 3 — Chunking Parameters

```python
CHUNK_SIZE = 500       # words
OVERLAP = 100          # words
```

Cell 4 — Chunk Generator

```python
CHUNK_SIZE = 500
OVERLAP = 100

chunks = []

chunk_id = 1

current_words = []
current_pages = []

for page in pages:

    words = page["text"].split()

    current_words.extend(words)

    current_pages.append(page["page"])

    while len(current_words) >= CHUNK_SIZE:

        chunk_words = current_words[:CHUNK_SIZE]

        chunks.append({

            "chunk_id": chunk_id,

            "start_page": current_pages[0],

            "end_page": page["page"],

            "text": " ".join(chunk_words),

            "word_count": len(chunk_words)

        })

        chunk_id += 1

        # Keep overlap
        current_words = current_words[CHUNK_SIZE - OVERLAP:]

        # Keep only the current page as the start reference for the next chunk.
        # We'll refine page mapping later if needed.
        current_pages = [page["page"]]

# Save any remaining words as the last chunk
if current_words:

    chunks.append({

        "chunk_id": chunk_id,

        "start_page": current_pages[0],

        "end_page": current_pages[-1],

        "text": " ".join(current_words),

        "word_count": len(current_words)

    })
```

Cell 5 — Verify

```python
print("=" * 60)

print("Total Chunks :", len(chunks))

word_counts = [c["word_count"] for c in chunks]

print("Minimum Words :", min(word_counts))
print("Maximum Words :", max(word_counts))
print("Average Words :", round(sum(word_counts) / len(word_counts), 2))

print("=" * 60)

print(chunks[0])
```

Cell 6 — Last Chunk

```python
print("="*60)
print("Last Chunk\n")
print(chunks[-1])
```

Cell 7 — Save Chunks

```python
OUTPUT_FILE = "../data/chunks.json"

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(chunks, f, indent=4, ensure_ascii=False)

print("Saved Successfully")

print(OUTPUT_FILE)
```

Cell 8 — Check File Size

```python
size = os.path.getsize(OUTPUT_FILE) / (1024 * 1024)

print(f"Chunk File Size : {size:.2f} MB")
```
